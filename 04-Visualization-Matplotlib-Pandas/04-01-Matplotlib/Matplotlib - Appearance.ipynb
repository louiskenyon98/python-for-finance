```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

#Adjust colour and line style
#Linewidth can be shortened to lw, as can other params
#Marker is used to mark discrete data points
ax.plot(x,y,color='purple',linewidth=3,alpha=0.5,linestyle='--',marker='o',markersize=20,markerfacecolor='yellow')
```

```python
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=2,alpha=0.5,ls='--')

#Set axis limits, line appears not curved as we do not have enough data points for this level of zoom
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
