```python
import matplotlib.pyplot as plt
```

```python
#Allow vewing of mpl within Jupyter
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
#without jupyter notebook you must use plt.show()
```

```python
#Multi plots on same canvas(rows,cols,plotnumber)
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented
```

```python
#Create figure object
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
#tuple unpacking
fig,axes = plt.subplots(nrows=1,ncols=2)

#axes is an object and its contents are accessible as normal
axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

#prevents overlapping
plt.tight_layout()
```
