```python
#Histogram
df1['A'].hist(bins=30)
```

```python
#Area plot, shows area of values
df2.plot.area()
```

```python
#Bar plot (must have categorical value as index)
df2.plot.bar(stacked=True)
```

```python
#Colour can be set using c parameter, therefore making a 3d plot
df1.plot.scatter(x='A',y='B',c='C')
```

```python
#Add colourmap to above
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

```python
#Boxplot
df2.plot.box()
```

```python
#Plotting bivariate data with a hex plot
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
df.plot.hexbin(x='a',y='b',gridsize=25)
```
